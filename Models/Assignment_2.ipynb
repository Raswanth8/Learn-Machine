### Data Visualization

```python
df = df.groupby('who')['fare'].sum().to_frame().reset_index()
```

```python
font1 = {'family':'serif','color':'green','size': 15}
font2 = {'family':'serif','color':'blue','size': 15}
font3 = {'family':'serif','color':'blue','size': 18}
plt.title("Bar Horizontal",fontdict = font3)
plt.xlabel("x",fontdict = font1)
plt.ylabel("y",fontdict = font2)
plt.barh(df['who'],df['fare'],color=['blue','green','red'])
plt.show()
```

```python
sns.barplot(x='fare',y='who',data=df,palette='Blues')
plt.title("Bar Horizontal",fontdict = font3)
plt.xlabel("x",fontdict = font1)
plt.ylabel("y",fontdict = font2)
plt.show()
```

```python
font1 = {'family':'serif','color':'green','size': 15}
font2 = {'family':'serif','color':'blue','size': 15}
font3 = {'family':'serif','color':'blue','size': 18}
plt.title("Bar ",fontdict = font3)
plt.xlabel("x",fontdict = font1)
plt.ylabel("y",fontdict = font2)
plt.bar(df['who'],df['fare'],color=['blue','green','red'])
plt.show()
```

```python
sns.barplot(x='who',y='fare',data=df,palette='Blues')
plt.title("Bar ",fontdict = font3)
plt.xlabel("x",fontdict = font1)
plt.ylabel("y",fontdict = font2)
plt.show()
```

```python
df = sns.load_dataset("titanic")
df_pivot = pd.pivot_table(df,values ='fare',index='who',columns='class',aggfunc=np.mean)
ax= df_pivot.plot(kind='bar',alpha =0.8)
plt.title("Grouped Bar ",fontdict = font3)
plt.xlabel("x",fontdict = font1)
plt.ylabel("y",fontdict = font2)
```

```python
sns.barplot(x='who',y='fare',hue='class',data=df,palette='Blues')
plt.title("Grouped Bar ",fontdict = font3)
plt.xlabel("x",fontdict = font1)
plt.ylabel("y",fontdict = font2)
plt.show()
```
