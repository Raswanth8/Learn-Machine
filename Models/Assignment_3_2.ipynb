```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy as sci

import matplotlib.pyplot as plt
```

```python
def estimate_coefficients(x, y):
    m = x.shape[0]
    mean_x, mean_y = np.mean(x), np.mean(y)

    sum_cross_deviation_xy, sum_squared_deviations_x = 0, 0

    for i in range(m):
        sum_cross_deviation_xy += (x[i] - mean_x) * (y[i] - mean_y)
        sum_squared_deviations_x += (x[i] - mean_x) ** 2
        
    beta1 = sum_cross_deviation_xy / sum_squared_deviations_x
    beta0 = mean_y - beta1 * mean_x
    print('Slope = {} and Intercept = {}'.format(beta1, beta0))
    print('y = x({}) + {}'.format(beta1, beta0))
    
    
    plt.figure(figsize=(10,6))
    plt.title('beta0 = {} , beta1 = {}'.format(beta0, beta1))
    plt.scatter(x,y, marker='o', color='blue')
    plt.plot(x, np.dot(x, beta1), 'r')
    return beta0, beta1
```
