```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("./dataset/train.csv")
```

```python
test = pd.read_csv("./dataset/test.csv")
```

```python
train
```

```python
df = train
df = df.append(test)
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#label encoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['owns_car'] = label_encoder.fit_transform(df['owns_car'].astype(str))
df['owns_house'] = label_encoder.fit_transform(df['owns_house'].astype(str))
df['gender'] = label_encoder.fit_transform(df['gender'].astype(str))
df.sample(20)
```

```python
df['owns_car'].unique()
```

```python
df['no_of_children'] = df['no_of_children'].fillna(df['no_of_children'].mean())
```

```python
df['no_of_days_employed'] = df['no_of_days_employed'].fillna(df['no_of_days_employed'].mean())
```

```python
df['yearly_debt_payments'] = df['yearly_debt_payments'].fillna(df['yearly_debt_payments'].mean())
```

```python
df['migrant_worker'] = df['yearly_debt_payments'].fillna(df['yearly_debt_payments'].mode())
df['total_family_members'] = df['yearly_debt_payments'].fillna(df['yearly_debt_payments'].mode())
```

```python
df['credit_score'] = df['credit_score'].fillna(df['credit_score'].mean())
df['credit_score'] = df['credit_score'].apply(np.ceil)
```

```python
df['name'] = label_encoder.fit_transform(df['name'])
df['occupation_type'] = label_encoder.fit_transform(df['occupation_type'])
```

```python
df
```

```python
df['credit_card_default'] = pd.get_dummies(df['credit_card_default'])
```

```python
df.isna().sum()
```

```python
df
```

```python
new_df = df.drop(['customer_id'],axis = 1)
```

```python
new_df.info()
```

```python
X = new_df.drop(['credit_card_default'],axis = 1)
X 
```

```python
y = new_df.credit_card_default
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

### LR

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', max_iter=150)
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

## Decision Tree

```python
from sklearn import tree
DT = tree.DecisionTreeClassifier(criterion = "entropy")
DT.fit(X_train, y_train)
```

```python
DT.score(X_train, y_train)
```

```python
y_pred_train = DT.predict(X_train)
```

```python
y_pred_test = DT.predict(X_test)
```

```python
y_pred_test
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = pd.DataFrame(confusion_matrix(y_test, y_pred_test))
plt.figure(figsize=(10,7))
sns.heatmap(confusion_matrix,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
